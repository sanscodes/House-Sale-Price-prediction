# Feature Engineering

```python
X = df2.copy()
y = X.pop("SalePrice")

# Label encoding for categoricals
for colname in X.select_dtypes("object"):
    X[colname], _ = X[colname].factorize()

# All discrete features should now have integer dtypes (double-check this before using MI!)
discrete_features = X.dtypes == int
```

```python
from sklearn.feature_selection import mutual_info_regression

def make_mi_scores(X, y, discrete_features):
    mi_scores = mutual_info_regression(X, y)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(X, y, discrete_features)
mi_scores[::3]
```

```python
data_corr = df2.select_dtypes(include=[np.number])
data_corr.head()
```

#Feature Selection

```python
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")


plt.figure(dpi=100, figsize=(8, 15))
plot_mi_scores(mi_scores)
```

#### REGRESSION

#LINEAR REGRESSION 
Training a Linear Regression model

```python
B = df2[['OverallQual', 'GrLivArea','GarageArea','GarageCars']] # these features affecting the most
b= df2['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
B_train, B_test, b_train, b_test = train_test_split(B, b, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(B_train,b_train)
```

```python
predictions = lm.predict(B_test)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(b_test, predictions))
print('MSE:', metrics.mean_squared_error(b_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(b_test, predictions)))
```

```python
from sklearn.metrics import r2_score
```

```python
score = {'Model Name':['Best Score', 'Test Score']}
```

```python
score = r2_score(b_test,predictions)
score
```
